# Repeat pandas

##### handling missing data

```python
import pandas as pd

url = "https://worldpopulationreview.com/country-rankings/most-educated-countries"

df_educ = pd.read_html(url)[2]
df_educ.head()
```

```python
df_educ.info()
```

```python
#replace "-" to null value: pd.NA or np.nan
#choose one of them to represent null values, for instance, column with np.nan can be used in math calculation, but it's not true for pd.NA
import numpy as np

df_educ = df_educ.replace("-", np.nan)


```

```python
df_educ.isnull().head()
```

```python
df_educ.isnull().sum()
```

```python
#clean Upper "Secondary Total 2022" column
df_educ["Upper Secondary Total 2022"] =df_educ["Upper Secondary Total 2022"].replace(r'[^0-9]', "", regex=True) #r is for raw string, ^ is for except
```

```python
df_educ["Upper Secondary Total 2022"]
```

```python
df_educ["Upper Secondary Total 2022"] = df_educ["Upper Secondary Total 2022"].astype(float)
```

```python
df_educ["Upper Secondary Total 2022"]
```

##### merge and replacing missing value

```python
url = "https://worldpopulationreview.com/country-rankings/list-of-countries-by-continent"

df_continent = pd.read_html(url)[-1]
df_continent.head()
```

```python
df_merged = pd.merge(df_educ,
                     df_continent[["Country", "Continent"]],
                     on="Country",
                     how="left"
                     )
df_merged.head()
```

```python
df_merged["Continent"].value_counts()
```

```python
df_merged.groupby("Continent")["Upper Secondary Total 2022"].mean()
```

##### we can use groupby() and transform() to replace missing values --> groupwise operation

```python
df_merged["Upper Secondary Total 2022"] = df_merged.groupby("Continent")["Upper Secondary Total 2022"].transform(lambda x: x.fillna(x.mean()))
```

```python
df_merged.head()
```

##### export df

```python
df_merged.to_csv("test.csv")
```
